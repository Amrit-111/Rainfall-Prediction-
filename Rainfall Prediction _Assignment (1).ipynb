```python
# Importing the libraries

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
from sklearn.linear_model import LinearRegression
```

```python
data =pd.read_csv("austin_weather_final.csv")
data
```

```python
X = data.drop(['PrecipitationSumInches'], axis =1 )
```

```python
Y = data['PrecipitationSumInches']
```

```python
#reshaping it into 2nd vector

Y = Y.values.reshape(-1,1)
```

```python
Y
```

```python
day_index = 798
days= [ i for i in range(Y.size)]
```

```python
# Intialize the linear regression classifier 

clf = LinearRegression()
```

```python
clf.fit(X,Y)
```

```python
Y = Y.astype(float).flatten()
```

```python
# plot a graph

print("The Precipitation trend graph:")

plt.figure(figsize=(10, 5))
plt.scatter(days, Y, color='g', alpha=0.6, label="Precipitation")
plt.scatter(days[day_index], Y[day_index], color='r', label=f"Day {day_index}")
plt.title("Precipitation Level Over Days")
plt.xlabel("Days")
plt.ylabel("Precipitation (inches)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
x_vis = X[['TempAvgF', 'DewPointAvgF', 'HumidityAvgPercent', 'SeaLevelPressureAvgInches', 'VisibilityAvgMiles', 'WindAvgMPH']]

```

```python
print("The precipitation vs attributes trend graph:")

plt.figure(figsize=(14, 10))  

for i, col in enumerate(x_vis.columns):
    plt.subplot(3, 2, i + 1)
    plt.scatter(days[:100], x_vis[col][:100], color='g', alpha=0.6)
    # Highlight day_index only if it's within the first 100
    if day_index < 100:
        plt.scatter(days[day_index], x_vis[col][day_index], color='r')
    plt.title(col)
    plt.xlabel("Days")
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```
